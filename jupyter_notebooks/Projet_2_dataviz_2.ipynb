```python
# importer les bibliothèques importantes
import pandas as pd
import numpy as np
import datetime
from datetime import timedelta
import plotly.express as px
```

```python
# importer le dataframe

films_db = pd.read_csv("C:/Users/tit_r/Desktop/Formation Data Analyst/Projets/Projet 2/Dataset/films_db.csv")
```

```python
# faire un split des genres pour pouvoir faire un explode par la suite

films_db['Genres'] = films_db['Genres'].str.split(',')
```

```python
# faire un explode des genres pour les utiliser dans des visuels

df_genres = films_db.explode('Genres')
```

```python
# faire un split des acteurs pour pouvoir faire un explode

films_db['liste_acteurs'] = films_db['liste_acteurs'].str.split(',')
```

```python
# faire un explode des acteurs pour utiliser dans les visuels

df_acteurs = films_db.explode('liste_acteurs')
```

```python
df_acteurs.info()
```

```python
# faire un camembert qui montre la part des 20 acteurs les plus présents

# compter l'occurence de chaque acteur
count_1 = df_acteurs['liste_acteurs'].value_counts()

# calculer le pourcentage de chaque acteurs
pourcent = count_1 / count_1.sum() * 100

# trier les valeurs par ordre décroissant et sélectionner les 20 premières
pourcent = pourcent.sort_values(ascending=False)[:20]

# créer piechart
fig_11 = px.pie(values=pourcent.values, names=pourcent.index)

fig_11.update_layout(title='Part des 20 acteurs les plus populaires')

fig_11.show()
```

```python
# histogramme des acteurs les plus présents

# grouper les données par acteur et compter le nombre d'occurrences
df_acteurs_count = df_acteurs.groupby('liste_acteurs')['Titre'].count()

# trier les valeurs et sélectionner les 20 meilleurs
df_acteurs_count = df_acteurs_count.sort_values(ascending=False)[:20]


fig_12 = px.bar(df_acteurs_count, x=df_acteurs_count.index, y=df_acteurs_count.values, text_auto=True)

fig_12.update_layout(title='Les 20 acteurs les plus présents', xaxis_title='Acteurs',
                 yaxis_title='Nombre de films', bargap=0.2)



fig_12.show()
```

```python
#  faire histogramme des réalisateurs les plus présents

# grouper les données par acteur et compter le nombre d'occurrences
df_real_count = films_db.groupby('Realisateur')['Titre'].count()

# trier les valeurs et sélectionner les 20 meilleurs
df_real_count = df_real_count.sort_values(ascending=False)[:20]


fig_13=px.histogram(df_real_count, x=df_real_count.index, y=df_real_count.values, text_auto=True)

fig_13.update_layout(title='Les 20 réalisateurs les plus présents', xaxis_title='Réalisateurs',
                 yaxis_title='Nombre de films', bargap=0.2)

fig_13.show()
```
